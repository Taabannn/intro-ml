# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.ylabel('funny numbers')
plt.show()
```

```python
plt.plot([7, 5, 9, 1], [-9, 3, 4, -1], 'ro')
```

```python
plt.plot([7, 5, 9, 1], [-9, 3, 4, -1], 'b+')
```

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, t**2, 'r--')
```

```python
plt.plot(t, t, 'r-.', t, t**2, 'bs', t, t**3, 'g^')
```

```python
data = {
    'a': np.arange(500),
    'b': np.random.randint(0, 500, 500),
    'c': np.random.randn(500)
}
```

```python
plt.scatter(data['a'], data['b'])
```

```python
plt.scatter(data['a'], data['c'])
```
