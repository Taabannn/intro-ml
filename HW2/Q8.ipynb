```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statistics import mean
```

# <span style="color:purple">Question 8</span>

```python
df=pd.read_csv("Data/ted_main.csv")
df.head()
```

```python
sns.kdeplot(df['duration'])
```

```python
X = df['duration'].values[:, np.newaxis]
h = 10

import math
num_in_window = sum(abs(xi - X) < (h/2) for xi in X)
#print(num_in_window)
n = len(X)
xd = np.linspace(-1000, 6000, 5000)
density = (1/(n*(h**2))) * sum(num*(1/math.sqrt(2*np.pi*h))*np.exp(-(xd-xi)**2/(2*(h)**2)) for xi, num in zip(X, num_in_window))
```

```python
plt.title('parzen window estimator wih h=10')
sns.kdeplot(df['duration'])
plt.plot(xd, density, label='estimated')
plt.xlabel('duration')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.neighbors import KernelDensity
kde20 = KernelDensity(kernel='gaussian', bandwidth=20)
kde20.fit(X)
kde50 = KernelDensity(kernel='gaussian', bandwidth=50)
kde50.fit(X)
kde100 = KernelDensity(kernel='gaussian', bandwidth=100)
kde100.fit(X)
kde250 = KernelDensity(kernel='gaussian', bandwidth=250)
kde250.fit(X)
```

```python
n = np.array(list(range(0, 2001)))[:, np.newaxis]

plt.plot(n, np.exp(kde20.score_samples(n)), label='20-band')
plt.plot(n, np.exp(kde50.score_samples(n)), label='50-band')
plt.plot(n, np.exp(kde100.score_samples(n)), label='100-band')
plt.plot(n, np.exp(kde250.score_samples(n)), label='250-band')
sns.kdeplot(df['duration'])
plt.legend()
plt.grid()
plt.title("TKDE Estimates by Bandwidth (Gaussian)")
```
